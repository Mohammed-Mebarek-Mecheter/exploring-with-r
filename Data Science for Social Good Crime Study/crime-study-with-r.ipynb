# Descriptive Statistics

```r
# Calculate counts and frequencies for calls
calls_summary <- calls_data %>%
  count(Original.Crime.Type.Name) %>%
  arrange(desc(n)) %>%
  mutate(frequency = n / sum(n) * 100)

# Calculate counts and frequencies for incidents
incidents_summary <- incidents_data %>%
  count(Category) %>%
  arrange(desc(n)) %>%
  mutate(frequency = n / sum(n) * 100)
```

```r
# Visualize the top 15 crime categories in calls
ggplot(data = head(calls_summary, 15), aes(x = reorder(Original.Crime.Type.Name, n), y = n)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  xlab("Crime Type") +
  ylab("Count") +
  ggtitle("Top 15 Calls for Service by Crime Type")

# Visualize the top 15 crime categories in incidents
ggplot(data = head(incidents_summary, 15), aes(x = reorder(Category, n), y = n)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  xlab("Crime Type") +
  ylab("Count") +
  ggtitle("Top 15 Police Incidents by Crime Type")
```

```r
calls_stats <- calls_summary %>%
  summarize(
    mean_count = mean(n),
    median_count = median(n),
    sd_count = sd(n)
  )

# Calculate additional descriptive statistics for incidents
incidents_stats <- incidents_summary %>%
  summarize(
    mean_count = mean(n),
    median_count = median(n),
    sd_count = sd(n)
  )

# Print the additional descriptive statistics
print("Descriptive Statistics for Calls:")
print(calls_stats)

print("Descriptive Statistics for Incidents:")
print(incidents_stats)
```

# Crime Types

```r
# Summarize the frequency of different crime types
crime_summary <- incidents_data %>%
  group_by(Category) %>%
  summarise(count = n()) %>%
  arrange(desc(count))

# Summarize crime types by neighborhoods
crime_by_neighborhood <- incidents_data %>%
  group_by(Analysis.Neighborhoods.2.2, Category) %>%
  summarise(count = n()) %>%
  arrange(desc(count))

# Summarize crime types by districts
crime_by_district <- incidents_data %>%
  group_by(PdDistrict, Category) %>%
  summarise(count = n()) %>%
  arrange(desc(count))
```

```r
# Bar plot of overall crime types
ggplot(crime_summary, aes(x = reorder(Category, -count), y = count)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs(title = "Frequency of Different Crime Types in San Francisco",
       x = "Crime Type", y = "Count")
```

```r
# Bar plot of crime types by neighborhood
ggplot(crime_by_neighborhood, aes(x = reorder(Analysis.Neighborhoods.2.2, -count), y = count, fill = Category)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs(title = "Crime Types by Neighborhood in San Francisco",
       x = "Neighborhood", y = "Count")
```

```r
# Bar plot of crime types by district
ggplot(crime_by_district, aes(x = reorder(PdDistrict, -count), y = count, fill = Category)) +
  geom_bar(stat = "identity") +
  coord_flip() +
  labs(title = "Crime Types by District in San Francisco",
       x = "Police District", y = "Count")
```

# Temporal Analysis

```r
# Convert date columns to Date format
calls_data$Call.Date <- as.Date(calls_data$Call.Date, format="%Y-%m-%d")
incidents_data$Date <- as.Date(incidents_data$Date, format="%Y-%m-%d")
```

```r
# Aggregate daily counts for calls
daily_calls <- calls_data %>%
  count(Call.Date) %>%
  rename(Date = Call.Date, Count = n)

# Aggregate daily counts for incidents
daily_incidents <- incidents_data %>%
  count(Date) %>%
  rename(Count = n)
```

```r
# Aggregate weekly counts for calls
weekly_calls <- calls_data %>%
  mutate(Week = floor_date(Call.Date, "week")) %>%
  count(Week) %>%
  rename(Date = Week, Count = n)

# Aggregate weekly counts for incidents
weekly_incidents <- incidents_data %>%
  mutate(Week = floor_date(Date, "week")) %>%
  count(Week) %>%
  rename(Date = Week, Count = n)
```

```r
# Aggregate monthly counts for calls
monthly_calls <- calls_data %>%
  mutate(Month = floor_date(Call.Date, "month")) %>%
  count(Month) %>%
  rename(Date = Month, Count = n)

# Aggregate monthly counts for incidents
monthly_incidents <- incidents_data %>%
  mutate(Month = floor_date(Date, "month")) %>%
  count(Month) %>%
  rename(Date = Month, Count = n)
```

```r
# Aggregate yearly counts for calls
yearly_calls <- calls_data %>%
  mutate(Year = floor_date(Call.Date, "year")) %>%
  count(Year) %>%
  rename(Date = Year, Count = n)

# Aggregate yearly counts for incidents
yearly_incidents <- incidents_data %>%
  mutate(Year = floor_date(Date, "year")) %>%
  count(Year) %>%
  rename(Date = Year, Count = n)
```

```r
# Plot daily trends for calls
ggplot(daily_calls, aes(x = Date, y = Count)) +
  geom_line(color = "blue") +
  labs(title = "Daily Crime Trends (Calls for Service)",
       x = "Date", y = "Count")

# Plot daily trends for incidents
ggplot(daily_incidents, aes(x = Date, y = Count)) +
  geom_line(color = "red") +
  labs(title = "Daily Crime Trends (Incidents)",
       x = "Date", y = "Count")
```

```r
# Plot weekly trends for calls
ggplot(weekly_calls, aes(x = Date, y = Count)) +
  geom_line(color = "blue") +
  labs(title = "Weekly Crime Trends (Calls for Service)",
       x = "Date", y = "Count")

# Plot weekly trends for incidents
ggplot(weekly_incidents, aes(x = Date, y = Count)) +
  geom_line(color = "red") +
  labs(title = "Weekly Crime Trends (Incidents)",
       x = "Date", y = "Count")
```

```r
# Plot monthly trends for calls
ggplot(monthly_calls, aes(x = Date, y = Count)) +
  geom_line(color = "blue") +
  labs(title = "Monthly Crime Trends (Calls for Service)",
       x = "Date", y = "Count")

# Plot monthly trends for incidents
ggplot(monthly_incidents, aes(x = Date, y = Count)) +
  geom_line(color = "red") +
  labs(title = "Monthly Crime Trends (Incidents)",
       x = "Date", y = "Count")
```

```r
# Plot yearly trends for calls
ggplot(yearly_calls, aes(x = Date, y = Count)) +
  geom_line(color = "blue") +
  labs(title = "Yearly Crime Trends (Calls for Service)",
       x = "Date", y = "Count")

# Plot yearly trends for incidents
ggplot(yearly_incidents, aes(x = Date, y = Count)) +
  geom_line(color = "red") +
  labs(title = "Yearly Crime Trends (Incidents)",
       x = "Date", y = "Count")
```
